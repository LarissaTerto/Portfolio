```python
from Bio import SeqIO

for record in SeqIO.parse("ecoli.fna", "fasta"):
    print(record.id, len(record.seq))
```

```python
def calculate_gc(seq):
    g = seq.count("G")
    c = seq.count("C")
    return (g + c) / len(seq) * 100

gc_values = []
for record in SeqIO.parse("ecoli.fna", "fasta"):
    gc = calculate_gc(str(record.seq))
    gc_values.append({"id": record.id, "gc_content": gc})
```

```python
import pandas as pd

df = pd.DataFrame(gc_values)
print(df.describe())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df["gc_content"], bins=30, kde=True)
plt.title("GC Content Distribution")
plt.show()
```
